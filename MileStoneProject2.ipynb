```python
# step 1

import random

suits = ('Hearts','Clubs','Diamonds','Spades')

ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')

values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12,'King':13,'Ace':14}

playing = True
```

```python
# creating a Card class
# step 2
class Card:

    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        

    def __str__(self):
        return self.rank + ' of ' + self.suit
    
```

```python
# creating a Deck of cards(52)
# step 3

class Deck():

    def __init__(self):

        # Creating a list of all cards
        self.deck = []

        for suit in suits:
            for rank in ranks:
                # creating the card using card class
                self.deck.append(Card(suit,rank))

    def __str__(self):
        st = ''
        for card in self.deck:
            st += '\n' + card.__str__()
        return "The Deck has:" + st
                
    def shuffle(self):

        random.shuffle(self.deck)

    def deal_one(self):
        return self.deck.pop()
```

```python
# step 4

class Hand:

    def __init__(self):
        self.cards = []    # stasrt with an empty list as we did in the Deck class
        self.value = 0    # start with zero value
        self.aces = 0     # add an attribute to keep track about aces

    def add_card(self,card):
        # card passed in 
        # From deck.deal() --> Single card(suit,rank)
        self.cards.append(card)
        self.value += values[card.rank]

    def adjust_for_ace(self):
        if self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
        
        
```

```python
# step 5

class Chips:

    def __init__(self):
        self.total = 100
        self.bet = 0

    def win_bet(self):
        self.total += self.bet

    def lose_bet(self):
        self.total -= self.bet
```

```python
# step 6 
# taking bets

def take_bet(chips):
    while True:
        try:
            chips.bet = int(input('Enter the bet amount: '))
        except ValueError:
            print('Bet must be an integer, try again', chips.total)
        else:
            if chips.bet > chips.total:
                print('Bet exceed the total funds')
            else:
                break
```

```python
# step 7
# creating hit function

def hit(deck,hand):
    hand.add_card(deck.deal_one())
    hand.adjust_for_ace()
```

```python
# step 8

def hit_or_stand(deck,hand):
    global playing
    x = input('Choose hit or Stand, Enter h or s')

    if x[0].lower() == 'h':
        hit(deck,hand)
        return
    elif x[0].lower() == 's':
        print("player stands, Dealer is playing.")
        playing = False
        return
    else:
        print("Error, try again")
        return hit_or_stand(deck,hand)
```

```python
# step 9
# creating a function to display the cards

def show_some(player,dealer):
    print("\nDealer's hand:")
    print('< Card Hidden >')
    print('',dealer.cards[1])
    print(" \nPlayer's hand:",*player.cards,sep = '\n')

def show_all(player,dealer):
    print("\nDealer's hand:",*dealer.cards,sep = "\n")
    print("Dealer's hand:",dealer.value)
    print("\nPlayer's hand:",*player.cards,sep = "\n")
    print("Player's hand:",player.value)
    
```

```python
# step 10
# functions to handle end of the game scenerio

def player_burst(player,dealer,chips):
    print('Player bursts')
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print('Player wins')
    chips.win_bet()

def dealer_burst(player,dealer,chips):
    print('Dealer bursts')
    chips.lose_bet()

def dealer_wins(player,dealer,chips):
    print('Dealer wins')
    chips.win_bet()

def push(player,dealer):
    print("Dealer and Player tie, it's a push")
```

```python
# Final step
# game on 

while True:

    print("Welcome to BlackJack! Get close to 21 as you can without going over!\n\
    Dealer hit's until she reaches 17. Aces count as 1 or 11")

    deck = Deck()
    deck.shuffle()

    # Dealing two cards for player
    player_hand = Hand()
    player_hand.add_card(deck.deal_one())
    player_hand.add_card(deck.deal_one())


    # Dealing two cards for Dealer
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal_one())
    dealer_hand.add_card(deck.deal_one())

    # Allocating chips for bet for the player
    player_chips = Chips()

    # placing a bet
    take_bet(player_chips)

    # showing the cards of player and dealer
    show_some(player_hand,dealer_hand)

    # started playing
    while playing:

        # Making a choice of hit or stand for the player
        hit_or_stand(deck,player_hand)

        # showing cards afer the hit or stand
        show_some(player_hand,dealer_hand)

        if player_hand.value > 21:
            player_burst(player_hand,dealer_hand,player_chips)
            break

    if player_hand.value < 21:

        while dealer_hand.value < 17:
            hit(deck,dealer_hand)

        show_all(player_hand,dealer_hand)

        if dealer_hand.value > 21:
            dealer_burst(player_hand,dealer_hand,player_chips)
            break
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)
            break
        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)
            break
        else:
            push(player_hand,dealer_hand)
            break

    print("\nPlayer's winnig's start at ",player_chips.total)

    new_game = input("Would you like to play another game? Enter yes or no")

    if new_game[0].lower() == 'y':
        playing = True
        continue
    else:
        print("Thank you for playing!")
        break
```
